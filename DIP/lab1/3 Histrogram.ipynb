```python
# Histrogram  https://www.youtube.com/watch?v=OKz2wedpDcM
import cv2
import matplotlib.pyplot as plt
import numpy as np

img  = cv2.imread("image3_grascale/hitlar.JPG") 

img_H = img.shape[0]
img_W = img.shape[1]

Histrogram = np.zeros([255], np.int32)

for x in range(0, img_H):
    for y in range(0, img_W):
        Histrogram[img[x,y]] +=1
        
for i in range(0, 255):
    print("Histrogram i :", i, Histrogram[i])


def Plot_Histrogram(img1):
    plt.figure()
    plt.title("GrayScale Histogram")

    plt.xlabel("Intensity Lavel")
    plt.ylabel("Intensity Frequency")

    plt.xlim([0, 255])

    plt.plot(img1)
    plt.savefig("H_G.jpg")

Plot_Histrogram(Histrogram)
```
