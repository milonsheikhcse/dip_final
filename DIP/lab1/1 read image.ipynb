```python
# read image 

import cv2
import matplotlib.pyplot as plt

img  = cv2.imread("image/hitlar.JPG") 
#rgb
b,g,r = cv2.split(img)
img = cv2.merge((r,g,b))
plt.imshow(img)

# write image
filename = 'Hitlar.JPG'
cv2.imwrite(filename, img)
```
